```python
confirmed_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
death_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv"
```

```python
df_conf = pd.read_csv(confirmed_url, index_col=0).reset_index()
df_conf.head()
```

```python
df_conf[df_conf['Province/State'] == 'Washington']
```

```python
df_deaths = pd.read_csv(death_url, index_col=0).reset_index()
df_deaths.head()
```

```python
us_conf = df_conf[df_conf['Country/Region'] == 'US'].drop('Country/Region',axis=1)
us_conf = us_conf.melt(id_vars=['Province/State','Lat','Long'], var_name='Date',value_name='Confirmed').set_index(['Province/State','Lat','Long','Date'])
us_conf.head()
```

```python
us_conf.shape
```

```python
us_deaths = df_deaths[df_deaths['Country/Region'] == 'US'].drop('Country/Region',axis=1)
us_deaths = us_deaths.melt(id_vars=['Province/State','Lat','Long'], var_name='Date',value_name='Deaths').set_index(['Province/State','Lat','Long','Date'])
us_deaths.head()
```

```python
us_deaths.shape
```

```python
us_covid19 = pd.concat([us_conf,us_deaths], axis=1).reset_index()
```

```python
us_covid19.head()
```
